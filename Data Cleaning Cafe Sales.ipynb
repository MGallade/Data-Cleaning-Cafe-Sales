```python
#import nececcasy libraries
import numpy as np
import pandas as pd
import plotly.express as px
import matplotlib as plt
import seaborn as sns
```

```python
#price List
price_dict={'Coffee':2,'Tea':1.5,'Sandwich':4,'Salad':5,'Cake':3,'Cookie':1,'Smoothie':4,'Juice':3}
```

```python
#import the csv file
df= pd.read_csv(r"C:\Users\hp\OneDrive\Desktop\datasets projects\cafe_sales.csv")
```

### Initial Checks

```python
df.head()
```

```python
df.info()
```

```python
#check duplicates(none present)
df.duplicated().sum()
```

```python
#check null values in each column
df.isnull().sum()
```

### For Column- Item

```python
#what all values are present
df['Item'].unique()
```

```python
#remove whitespaces if any
df['Item'] = df['Item'].str.strip().str.capitalize()
```

```python
#replace irrelevant entries
df['Item']=df['Item'].replace(['Unknown','Error'],pd.NA)
```

```python
#remove null values
df.dropna(subset=['Item'], inplace=True)
```

### For Column- Location

```python
#what all values are present
df['Location'].unique()
```

```python
#fill null values
df['Location']=df['Location'].fillna('Delivery')
```

```python
#replace irrelevant entries with Delivery
df['Location']=df['Location'].replace('UNKNOWN','Delivery')
df['Location']=df['Location'].replace('ERROR','Delivery')
```

### For Column- Quantity

```python
#what all values are present
df['Quantity'].unique()
```

```python
#convert to float values
df['Quantity'] = pd.to_numeric(df['Quantity'], errors='coerce')
```

```python
#remove null values
df.dropna(subset=['Quantity'], inplace=True)
```

### For Column- Price per unit

```python
#what all values are present
df['Price Per Unit'].unique()
```

```python
#turn to float values
df['Price Per Unit'] = pd.to_numeric(df['Price Per Unit'], errors='coerce')
```

```python
#map the prices from the prices at the beginning
df['Price Per Unit'] = df['Price Per Unit'].fillna(df['Item'].map(price_dict))
```

### For Column- Total Spent

```python
#turn to float values
df['Total Spent'] = pd.to_numeric(df['Total Spent'], errors='coerce')
```

```python
#total=Price*quantity
df['Total Spent']=df['Quantity']*df['Price Per Unit']
```

### For Column- Payment Method

```python
#what all values are present
df['Payment Method'].unique()
```

```python
#replace irrelevant entries with UPI, and replace Digital Wallet with Swiggy Dineout

df['Payment Method']=df['Payment Method'].replace(['UNKNOWN','ERROR'],pd.NA)
df['Payment Method']=df['Payment Method'].replace('Digital Wallet','Swiggy Dineout')
df['Payment Method']=df['Payment Method'].fillna('UPI')
```

### For Column- Transaction Date

```python
df['Transaction Date'] = pd.to_datetime(df['Transaction Date'], errors='coerce')
```

```python
df = df.dropna(subset=['Transaction Date'])
```
